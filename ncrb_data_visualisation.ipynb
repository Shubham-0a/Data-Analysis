# Getting data from the ncrb (national crime records beureu )

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
curr_dir = os.getcwd()
data_2017_df = pd.read_excel(curr_dir + '/data_2017.xlsx')
data_2018_df = pd.read_excel(curr_dir + '/data_2018.xlsx')
data_2020_df = pd.read_excel(curr_dir + '/data_2020.xlsx')
data_2019_df = pd.read_excel(curr_dir + '/data_2019.xlsx')

# removing all nan data

# 2017
data_2017_df = data_2017_df.drop(0)
data_2017_df = data_2017_df.drop(2)
data_2017_df = data_2017_df.drop(3)
data_2017_df = data_2017_df.dropna()
# 2018
data_2018_df = data_2018_df.drop(0)
data_2018_df = data_2018_df.drop(2)
data_2018_df = data_2018_df.drop(3)
data_2018_df = data_2018_df.dropna()
# 2019
data_2019_df = data_2019_df.drop(0)
data_2019_df = data_2019_df.drop(2)
data_2019_df = data_2019_df.drop(3)
data_2019_df = data_2019_df.dropna()
# 2020
data_2020_df = data_2020_df.drop(0)
data_2020_df = data_2020_df.drop(2)
data_2020_df = data_2020_df.drop(3)
data_2020_df = data_2020_df.dropna()
# resetting the DataFrame index
data_2017_df = data_2017_df.reset_index()
data_2018_df = data_2018_df.reset_index()
data_2019_df = data_2019_df.reset_index()
data_2020_df = data_2020_df.reset_index()

# modifying 2017 data
data_2017_df.columns
cyber2017_df = pd.DataFrame({})
cyber2017_df[data_2017_df['Unnamed: 1'][0]] = data_2017_df['Unnamed: 1'][1:]
cyber2017_df[data_2017_df['Unnamed: 2'][0]] = data_2017_df['Unnamed: 2'][1:]
cyber2017_df[data_2017_df['Unnamed: 3'][0]] = data_2017_df['Unnamed: 3'][1:]
cyber2017_df[data_2017_df['Unnamed: 4'][0]] = data_2017_df['Unnamed: 4'][1:]
cyber2017_df[data_2017_df['Unnamed: 5'][0]] = data_2017_df['Unnamed: 5'][1:]
cyber2017_df[data_2017_df['Unnamed: 6'][0]] = data_2017_df['Unnamed: 6'][1:]
cyber2017_df[data_2017_df['Unnamed: 7'][0]] = data_2017_df['Unnamed: 7'][1:]
cyber2017_df[data_2017_df['Unnamed: 8'][0]] = data_2017_df['Unnamed: 8'][1:]
cyber2017_df['Year'] = str(np.datetime64('2017'))
cyber2017_df.to_csv('2017_data.csv')

# modifying 2018 data
data_2018_df.columns
cyber2018_df = pd.DataFrame({})
cyber2018_df[data_2018_df['Unnamed: 1'][0]] = data_2018_df['Unnamed: 1'][1:]
cyber2018_df[data_2018_df['Unnamed: 2'][0]] = data_2018_df['Unnamed: 2'][1:]
cyber2018_df[data_2018_df['Unnamed: 3'][0]] = data_2018_df['Unnamed: 3'][1:]
cyber2018_df[data_2018_df['Unnamed: 4'][0]] = data_2018_df['Unnamed: 4'][1:]
cyber2018_df[data_2018_df['Unnamed: 5'][0]] = data_2018_df['Unnamed: 5'][1:]
cyber2018_df[data_2018_df['Unnamed: 6'][0]] = data_2018_df['Unnamed: 6'][1:]
cyber2018_df[data_2018_df['Unnamed: 7'][0]] = data_2018_df['Unnamed: 7'][1:]
cyber2018_df[data_2018_df['Unnamed: 8'][0]] = data_2018_df['Unnamed: 8'][1:]
cyber2018_df['Year'] = str(np.datetime64('2018'))
cyber2018_df.to_csv('2018_data.csv')

# modifying 2019 data
data_2019_df.columns
cyber2019_df = pd.DataFrame({})
cyber2019_df[data_2019_df['Unnamed: 1'][0]] = data_2019_df['Unnamed: 1'][1:]
cyber2019_df[data_2019_df['Unnamed: 2'][0]] = data_2019_df['Unnamed: 2'][1:]
cyber2019_df[data_2019_df['Unnamed: 3'][0]] = data_2019_df['Unnamed: 3'][1:]
cyber2019_df[data_2019_df['Unnamed: 4'][0]] = data_2019_df['Unnamed: 4'][1:]
cyber2019_df[data_2019_df['Unnamed: 5'][0]] = data_2019_df['Unnamed: 5'][1:]
cyber2019_df[data_2019_df['Unnamed: 6'][0]] = data_2019_df['Unnamed: 6'][1:]
cyber2019_df[data_2019_df['Unnamed: 7'][0]] = data_2019_df['Unnamed: 7'][1:]
cyber2019_df[data_2019_df['Unnamed: 8'][0]] = data_2019_df['Unnamed: 8'][1:]
cyber2019_df['Year'] = str(np.datetime64('2019'))
cyber2019_df.to_csv('2019_data.csv')

# modifying 2020 data
data_2020_df.columns
cyber2020_df = pd.DataFrame({})
cyber2020_df[data_2020_df['Unnamed: 1'][0]] = data_2020_df['Unnamed: 1'][1:]
cyber2020_df[data_2020_df['Unnamed: 2'][0]] = data_2020_df['Unnamed: 2'][1:]
cyber2020_df[data_2020_df['Unnamed: 3'][0]] = data_2020_df['Unnamed: 3'][1:]
cyber2020_df[data_2020_df['Unnamed: 4'][0]] = data_2020_df['Unnamed: 4'][1:]
cyber2020_df[data_2020_df['Unnamed: 5'][0]] = data_2020_df['Unnamed: 5'][1:]
cyber2020_df[data_2020_df['Unnamed: 6'][0]] = data_2020_df['Unnamed: 6'][1:]
cyber2020_df[data_2020_df['Unnamed: 7'][0]] = data_2020_df['Unnamed: 7'][1:]
cyber2020_df[data_2020_df['Unnamed: 8'][0]] = data_2020_df['Unnamed: 8'][1:]
cyber2020_df['Year'] = str(np.datetime64('2020'))
cyber2020_df.to_csv('2020_data.csv')
```
